```python
image_string=tf.io.read_file("F:/Projects/Neural Net/Galaxy Zoo/Data/images_training_rev1/New folder/119150.jpg")
image=tf.image.decode_jpeg(image_string,channels=3)
```

```python
def visualize(original, augmented):
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        original = sess.run(original)
        print(type(original))
        fig = plt.figure()
        plt.subplot(1,2,1)
        plt.title('Original image')
        plt.imshow(original)
        
        augmented = sess.run(augmented)
        plt.subplot(1,2,2)
        plt.title('Augmented image')
        plt.imshow(augmented)
```

```python
rand_crop = tf.image.random_crop(image,[300,300,3], seed = None)
```

```python
visualize(image, rand_crop)
```

```python
whitening = tf.image.per_image_standardization(image)
```

```python
visualize(image, whitening)
```

```python
visualize(image, whitening)
```

```python
crop = tf.image.central_crop(image, 224/424)
```

```python
visualize(image, crop)
```

```python
whitening = tf.image.per_image_standardization(crop)
visualize(crop, whitening)
```

```python
central_crop = tf.image.central_crop(image, 0.6)
central_crop = tf.image.resize(central_crop, [424,424], method='nearest')
visualize(image, central_crop)
```
